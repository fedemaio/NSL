## <span style="color:navy">  Exercise 9 - Travelling Salesman Problem</span>


The Travelling Salesman Problem represents a classical example of a Genetic Algorithm application.

Given a set of cities randomly distributed (on a 2D space), the aim of the salesman is visiting all the cities only once with the shortest possible round-tour.

Every possible tour is then classified on a <i>'fitness'</i> basis, related to its (square) lenght:

$L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2$

where $\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N$ are the positions of cities in the order they are visited and s.t. $\vec{x}_{N+1}=\vec{x}_1$.

## Genetic Algorithm

The Genetic Algorithm, as the name suggests, is based on the phenomen of the <i>natural selection</i> among the individuals of a spiece according to their ability to adapt. Generation after generation, only the most adaptable ones survive and reproduce.

The GA is then based on the following steps:

0. Create the **initial population** with random individuals;
1. Evaluate the **fitness** of each individual and so the probability to be selected.
2. Produce the **next generation**:
    a. select a parent;
    b. with probability $P_c$, **crossover**: a part of its genes will be merge with genes of another selected parent to produce an offspring.
    c. if crossover doesn't occur, then copy the parent in the next generation;
    d. with probability $P_m$, **mutation**: the offspring can face some unpredictable changes;
3. The last step is repeated until the new generation doesn't fullfil the fitness conditions.

## Exercise

In order to implement this idea in the TSP case, I have created the two following classes:
- <i>**Tour**</i>: a vector of cities with some methods implemented to make some operations more efficient  
- <i>**Population**</i>: it is instead a vector of tours, which is able to execute all the operations just mentioned (initialization, selection, crossover, mutation, sorting...).

The number of cities, their position configuration, the number steps, the number of chromosomes and the probabilities of crossover and mutation are all contained in an input file (input.dat) to be quickly changed to study some properties of the algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Circumference

### Best path as a function of steps of the algorithm

```python
data = np.loadtxt('averages_C.out')

step = data[:, 0]
best_length = data[:, 1]
mean_best_length = data[:, 2]

plt.figure(figsize=(10, 6))

plt.plot(step, best_length, label='Best Length', color='darkcyan')

plt.plot(step, mean_best_length, label='Best Length averaged on the best half of population', color='darkmagenta')

plt.title('Graph of Lengths at each Step')
plt.xlabel('Steps')
plt.ylabel('Length')

plt.legend()
plt.grid(True)
plt.show()
```

### Best path in cartesian coordinates

```python
data = np.loadtxt('coordinates_C.out')

x = data[:, 1]  
y = data[:, 2]  

plt.figure(figsize=(8, 6))

plt.plot(x, y, marker='o', linestyle='-', color='skyblue', markerfacecolor='darkmagenta', markersize=8, markeredgewidth=1.5, label='cities')

plt.scatter(x[0], y[0], s=200, color='red', edgecolor='black', zorder=10, marker='o', label='starting point')

plt.title('Coordinates plot of the best path')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
plt.grid(True)
plt.show()
```
